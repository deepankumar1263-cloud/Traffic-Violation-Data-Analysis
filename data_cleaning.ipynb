```python
from rapidfuzz import process, fuzz
import pandas as pd

make_list_lower = [m.lower() for m in make_list]

def standardize_make(val,og_list):
    val = str(val).lower()
    match = process.extractOne(val,og_list, scorer=fuzz.ratio, score_cutoff=50)
    
    if match:
        return match[0]
    return val

df["Make"] = df["Make"].apply(standardize_make,args=(make_list_lower,))
```

```python
def others(row,make_list):
    if row not in make_list and row!="not applicable":
        return("others")
    else:
        return(row)
    
df["Make"]=df["Make"].apply(others,args=(make_list_lower,))
df["Make"].value_counts()
```
